```python
# - 3.Create a python program to display multiple types of charts using matplotlib package
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y1 = [1,2,3,4,5]
y2 = [2,4,6,7,9]
plt.plot(x,y1)
plt.title("Single line plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# - Multiline plot
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("Multi line plot")
plt.xlabel("X")
plt.ylabel("y")
plt.show()

# - Bar chart
plt.bar(x,y1,color = "red")
plt.title("Bar chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# - Scatter plot
plt.scatter(x,y1)
plt.title("Scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# - 7.Demonstrate the following functions/methods which operates on tuples in Python with suitable examples: 
#             i) len( ) ii) count( ) iii) index( ) iv) sorted( ) v) min ( ) vi)max( ) 

# - len()
my_tuple = (1,2,3,4,5,6)
length = len(my_tuple)
print("Length of the tuple:",length)

# - count()
count1 = my_tuple.count(1)
print("Number of times 1 appears:",count1)

# - index()
indexvalue = my_tuple.index(2)
print("Index of two:",indexvalue)

# - sorted() The sorted() function returns a sorted list of the elements in the tuple.
sorted_tuple = sorted(my_tuple)
print("sorted tuple:",sorted_tuple)

# - min()
minimum = min(my_tuple)
print("The minimum value:",minimum)

# - max()
maximum = max(my_tuple)
print("The maximum value:",maximum)
```

```python
# - 8. Write a Python program to create bar plot from a DataFrame.
'''Sample Data Frame:
a b c d e
2 4,8,5,7,6
4 2,3,4,2,6
6 4,7,4,7,8
8 2,6,4,8,6
10 2,4,3,3,2
'''
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'a':[1,2,3,4,5],
    'b':[2,3,4,5,6]
}
df = pd.DataFrame(data)
df.plot(kind="bar")
plt.title("Bar plot")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

```python
# - 9.Demonstrate the following functions/methods which operates on sets in Python with suitable examples:
#          i) add( ) ii) update( ) iii) copy( ) iv) pop( ) v) remove( ) 

# - add( ) 
my_set = {1,2,3}
my_set.add(4)
print(my_set)

# - update( )
myset = {1,2,3}
myset.update([4,5,6])
print(myset)

# - copy()
mysets = {1,2,3}
newset = mysets.copy()
print(newset)

# - pop()
popped_element = mysets.pop()
print(popped_element)

# - Remove()
mysetss = {1,2,3}
mysetss.remove(1)
print(mysetss)
```

```python
# - 10. Write a Python program to create bar plot of scores by group and gender. 
'''Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
'''
import matplotlib.pyplot as plt
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]
plt.bar(groups, means_men, color='b', width=0.4, label='Men')
plt.bar(groups, means_women, color='r', width=0.4, label='Women', bottom=means_men)
plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')
plt.legend()
plt.show()
```

```python

# Example string with leading and trailing spaces
my_string = "   Hello, World!   "

# Using strip() to remove leading and trailing spaces
stripped_string = my_string.strip()
print("Stripped string:", stripped_string)
```

```python
# Example string with trailing spaces
my_string = "Hello, World!   "

# Using rstrip() to remove trailing spaces
rstrip_string = my_string.rstrip()
print("Right stripped string:", rstrip_string)
```
